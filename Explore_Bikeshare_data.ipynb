**in each city, what's month have most users use bike? what's month have minimun users use bike?**

```r
# Your solution code goes here
#a<- (ny$Start.Time)
#a[1]
#as.Date(a[1])
library(ggplot2)
#library(ggthemes)
#qplot(x=as.Date(Start.Time),data=ny, color = 'black', fill = '#099DD9')+
#xlab('Newyork - Bike use time') +
#ylab('Newyork - Number of user')+
#stat_bin( bins=30, binwidth=1)
ggplot(aes(x = as.Date(Start.Time)), data = ny) +
geom_histogram(binwidth = 30, color = 'black', fill = '#099DD9')+
xlab('Newyork - Bike use by month') +
ylab('Newyork - Number of user')
```

```r
ggplot(aes(x = as.Date(Start.Time)), data = wash) +
geom_histogram(binwidth = 30, color = 'black', fill = '#099DD9')+
xlab('Washington - Bike use by month') +
ylab('Washington - Number of user')
```
